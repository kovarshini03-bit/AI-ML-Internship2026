```python
#case conversion
n="double bunny"
print(n.upper())
print(n.lower())
print(n.capitalize())
print(n.title())
print(n.swapcase())
```

```python
#alignment and padding
n="rapmonster"
print(n.center(20,'*'))
print(n.ljust(10,'*'))
print(n.rjust(30,'*'))
print(n.zfill(5))
```

```python
#splitting and joining
bts="min yoongi"
print(bts.split())
print(bts.split("o"))
print(bts.join(["yoongi","min"]))
print(bts.rsplit("i"))
print(bts.partition(" "))
```
